(4) Matplotlib의 그래프 종류

```python
# 리스트 형식의 막대그래프
plt.bar(['사과','바나나','키위','방울토마토'], [10, 3, 6, 20])
plt.show()
```

```python
# 딕셔너리 형식의 막대그래프
dict1 ={'index':['사과','바나나','키위','방울토마토'],
        'data':[10, 3, 5, 20]}
lstColor = ['pink','b','skyblue','#e35f62']

plt.bar('index','data', data=dict1, color=lstColor)
plt.show()
```

```python
# 막대그래프
plt.barh('index','data', data=dict1, color=lstColor, height=0.7)
plt.show()
```

```python
# 파이 형식의 그래프
plt.axis('equal')
plt.pie('data', labels='index', data=dict1)
plt.show()
```

```python
# 설정 변경
plt.pie("data",labels="index", data = dict1, colors=lstColor, startangle=45, autopct="%.1f%%", explode=(0,0.1,0,0)) 
plt.show()
```

# 4단계 : 데이터 시각화

```python
# 한글폰트
# 음수설정
plt.rc('font', family = 'NanumGothic')
plt.rc('axes', unicode_minus = False)
```

```python
plt.rcParams['font.size']=14
plt.rcParams['lines.linewidth'] = 2
plt.figure(figsize=(30,8))
plt.title("서울시 2012년~2022년 기온분석", fontsize=24)
plt.plot(data.index, data['최고기온'], c='r', marker='^', label='최고기온')
plt.plot(data.index, data['최저기온'], c='b', marker='o', label='최저기온')
plt.xticks(rotation=70)
plt.legend()
plt.grid()
plt.show()
```

```python
lstColor = []
for i in range(len(data.index)):
    lstColor.append("red")
plt.figure(figsize=(40, 8))
plt.title("서울시 2012년 ~ 2022년도 기온데이터", fontsize=24)
data['최저기온'].plot(kind='bar', x= data.index)
data['최고기온'].plot(kind='bar', x= data.index, color=lstColor)
plt.xticks(rotation=60)
plt.grid()
plt.show()
```
